```python
# Standardize usernames
chat['username'] = chat['username'].str.upper()
chat.head()
```

```python
def preprocess_text(text):
    # Lowercase
    text = text.lower()
    # Remove punctuation and special characters
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    text = re.sub(r'http\S+', '', text)
    text = re.sub(r'[0-9]+','', text)
    text = re.sub(r'\s+',' ', text)
    text = re.sub(r'[^\w\s]|_', '', text)
    text = re.sub(r'([a-zA-Z])\1\1','\\1', text)



    text = text.replace('<Media omitted>', '').replace('This message was deleted', '').replace('\n', ' ').replace('video omitted', '').replace('image omitted', '').strip()

    # Tokenize
    tokens = word_tokenize(text)
    # Remove stop words
    tokens = [word for word in tokens if word not in stopwords.words('english')]
    # Lemmatize
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    return ' '.join(tokens)
```

```python
# Apply text preprocessing
chat['cleaned_message'] = chat['message'].apply(preprocess_text)
```

**Sentiment Analysis using VADER**

```python
# initializing NLTK sentiment analyzer

analyzer = SentimentIntensityAnalyzer()

# creating a get_sentiment function
def get_sentiment(text):
    scores = analyzer.polarity_scores(text)

    if scores['compound'] >= 0.05:
        sentiment = 'positive'
    elif scores['compound'] <= -0.05:
        sentiment = 'negative'
    else:
        sentiment = 'neutral'

    return sentiment


# applying the get_sentiment function into the dataframe

chat['sentiment'] = chat['message'].apply(get_sentiment)

print(chat)
```

```python
fig = go.Figure(data=go.Pie(labels=['Negative', 'Neutral', 'Positive'],
                            values=chat.groupby('sentiment').count()[['message']].reset_index()['message'],hole = 0.3))
fig.update_traces(hoverinfo='label+value')
```
